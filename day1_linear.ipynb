### 1. 선형리스트 (= 순차리스트)
- 배열
- 빈틈없이 붙어있는것이라고 생각하면 됨 
- 장점 : 공간절약(연결과 달리 링크가 없기때문), 전체를 접근하기에 빠름 
- 단점 : 데이터 삽입 / 삭제시 오버헤드가 발생될 수 있음 
- ex) 삽입/삭제가 필요 없는 데이터 = 시간순차적 데이터
    - 신문기사, 고전소설 연대문

- 선형리스트의 원리

````python
### 함수 선언부

### 전역 변수부
katok = ['다현', '정연', '쯔위', '사나', '지효'] # 선형리스트


### 메인코드부
print(katok)

## 1) 데이터 추가 : 모모자리 만들어주기 
katok.append(None) # 모모를 추가하기 위해 "마지막 자리"에 빈칸 생성 
katok[5] = '모모'
print('\n새로운데이터 추가 :\n', katok) # 모모가 삽입됨


## 2) 데이터 삽입 : 미나가 3등이 됨
# 1단계 : 맨 끝자리에 빈칸 준비 
katok.append(None)

# 2단계 : 3등까지 친구들을 오른쪽으로 한 칸 이동시키기 <반복>
katok[6] = katok[5] # 모모 --> 인덱스6에 5였던 모모 집어넣기
katok[5] = None # --> 인덱스 5는 비웠으니 none 입력
katok[5] = katok[4] # 지효 --> 빈 5등자리에 4등 집어넣기 
katok[4] = None # --> 4등자리는 빈석
katok[4] = katok[3] # 사나 --> 빈 4등자리에 3등이었던 사나 집어넣기 
katok[3] = None # --> 3등은 공석으로 만들기 

# 3 단계 : 빈칸(3등) 자리에 미나 삽입
katok[3] = '미나'
print('\n3등 삽입 :\n', katok)



## 3) 데이터 삭제 : 4등의 사나가 카톡을 차단함 
# 1단계 : 데이터 제거
katok[4] = None # 4등 자리 공석으로 만들기 

# 2단계 : 빈자리 없애기 위해 자리 앞당기기 
katok[4] = katok[5] # 4등자리에 5등 앉히기
katok[5] = None # 5등은 공석
katok[5] = katok[6] # 5등 자리에 6등 앉히기
katok[6] = None # 6등은 공석
# 3단계 : 빈칸 제거
del(katok[6]) # 남은 공석은 아예 제거하기 
print('\n4등 삭제 :\n', katok)
````

- 배열을 이용한 선형 리스트 생성
    - 함수생성하여 데이터 추가, 삽입, 제거 해보기

````python
### 변수 
katok = []


### 함수
## 1) 데이터 추가하는 함수
def add_data(friend) :
    # 1) 빈자리 생성
    katok.append(None) 
    # 2) 데이터 추가 
    klen = len(katok)
    katok[klen-1] = friend


## 2) 리스트 중간에 데이터 삽입하기
def insert_data(pos, friend) :
    # 1) 빈자리 생성
    katok.append(None)
    # 2) 데이터 공석 만들어주기
    klen = len(katok)
    for i in range(klen-1, pos, -1) :
        katok[i] = katok[i-1]
        katok[i-1] = None
    # 3) 데이터 삽입    
    katok[pos] = friend


## 3) 데이터 삭제 
def delete_data(pos) :
    # 1) 삭제할 데이터 자리 공석으로 만들어주기
    katok[pos] = None
    # 2) 뒷자리 당겨서 공석 체워주기  
    klen = len(katok)
    for i in range(pos, klen-1, 1) :
        katok[i] = katok[i+1]
        katok[i+1] = None
        
    # 3) 남는 공석 제거 
    del(katok[klen-1])
    
    
    
### 메인 
## 데이터 추가 
add_data('다현')
add_data('졍연')
add_data('쯔위')
add_data('사나')
add_data('지효')
print(katok)
add_data('모모') # 한 명 더 추가 
print(katok)
## 데이터 삽입 
insert_data(3, '미나')
print('\n데이터 중간에 삽입 \n:', katok)
## 데이터 제거 
delete_data(4)
print('\n데이터 제거됨 :\n', katok) # 사나가 제거됨을 알 수 있음 
````

#### 4. head, pre, current
- head : 첫 번째 노드를 가리킴 
- current : 지금 처리중인 노드
- pre : 현재 노드의 바로 앞 노드 
- 처음에는 모두 비어있으면 되므로 

```bash
    memory = []
    head, current, pre = None, None, None 
```
- 으로 처리하면됨

4-1. 데이터 연결

````python
### 전역변수
memory = []
head, pre, current = None, None, None
dataArray = ['수영', '서현', '태연', '유리', '윤아'] # 실제 삽입될 데이터 목록 # 실무에서는 쿼리된 결과

### 함수
class Node():
    def __init__(self) :
        self.data = None
        self.link = None
        
# head를 집어넣으면 데이터를 연결시켜주는 함수       
def printNodes(start) :
    current = start  # 처음에는 head와 current가 같음
    if current == None: # current에 Node가 있으면 돌아감
        return 
    print(current.data, end = ' ') # head의 데이터 출력 
    while current.link != None : # 만약 current.link가 None을 잡지 않는다면
        current = current.link # 앞 노드의 link를 갖게함
        print(current.data, end=' ') 
    print()
        


### 메인 
# head를 만들기 위한 코드 
node = Node()
node.data = dataArray[0] # 첫 번째 노드
head = node # 첫 노드는 head 
memory.append(node) # 첫 노드를 memory에 넣음 

# 나머지는 몇 백개가 될지 모르니 for문으로 생성
for data in dataArray[1:] : # 두 번째 노드부터 끝까지 데이터를 삽입하기 위함 (0번째는 head가 있음)
    pre = node # head를 잡고있기 위한 변수
    node = Node()
    node.data = data # 두 번째 노드 이후
    pre.link = node
    memory.append(node)
    
printNodes(head) # head를 집어넣으면 데이터를 연결시켜주는 함수 

````

4-3. 노드 삽입하는 함수 생성

````python
### 전역변수
memory = []
head, pre, current = None, None, None
dataArray = ['수영', '서현', '태연', '유리', '윤아'] # 실제 삽입될 데이터 목록 # 실무에서는 쿼리된 결과

### 함수
class Node():
    def __init__(self) :
        self.data = None
        self.link = None
        
# head를 집어넣으면 데이터를 연결시켜주는 함수       
def printNodes(start) :
    current = start  # 처음에는 head와 current가 같음
    if current == None: # current에 Node가 있으면 돌아감
        return 
    print(current.data, end = ' ') # head의 데이터 출력 
    while current.link != None : # current.link가 None을 잡지 않을 때까지 돌리기
        current = current.link # --> current = current.link = newNode 이니 결국 current가 다음 노드를 새로 갖게됐다고 생각하면 됨 
        print(current.data, end=' ') # --> 이 짓을 새로운 노드(current.link)가 None이 아닐때 까지 하면 됨
    print()
        

# 노드 삽입하는 함수
def insertNodes(findData, insertData) :
    global memory, head, pre, current
    # case 1 : head 앞에 노드 삽입하기 
    if (findData == head.data) :
        node = Node()
        node.data = insertData
        node.link = head # 새로운 노드의 link가 원래 head의 노드를 잡게하기
        head = node # 새로운 노드에 head 넘겨주기
        memory.append(node)
        return 
    # case 2 : 중간에 노드 삽입하기 
    current = head # 커렌트가 head에 있음을 확인 
    while (current.link != None) :
        pre = current # pre에 현재 노드 넘겨주기
        current = current.link # 현재 노드는 새로운 노드를 받기
        if current.data == findData :
            node = Node() # 삽입할 노드 생성
            node.data = insertData
            node.link = current
            pre.link = node
            memory.append(node) # 삽입된 노드 담기
            return
    # case 3 : 마지막에 노드 삽입하기 
    node = Node() # 마지막에 삽입할 노드 생성
    node.data = insertData
    current.link = node # 현재노드 링크가 새로운 노드 잡게하기
    return






### 메인 
# head를 만들기 위한 코드 
node = Node()
node.data = dataArray[0] # 첫 번째 노드
head = node # 첫 노드는 head 
memory.append(node) # 첫 노드를 memory에 넣음 

# 나머지는 몇 백개가 될지 모르니 for문으로 생성
for data in dataArray[1:] : # 두 번째 노드부터 끝까지 데이터를 삽입하기 위함 (0번째는 head가 있음)
    pre = node # pre : 앞 노드를 잡고있기 위한 변수
    node = Node() # 새로운 노드 생성
    node.data = data # 두 번째 노드에 데이터 삽입
    pre.link = node # 이전 노드의 링크가 새로운 노드를 잡게됨
    memory.append(node) # 새로 생긴 노드가 담기게됨

print('초기 상태')    
printNodes(head) # head를 집어넣으면 데이터를 연결시켜주는 함수 

# head에 새로운 node 삽입 
print('\nhead에 삽입 :')
insertNodes('수영', 'melody')
printNodes(head)

# 중간에 삽입하기 
print('\n중간에 삽입 :')
insertNodes('태연', 'amy')
printNodes(head)

# 마지막에 삽입하기 
print('\n마지막에 삽입 :')
insertNodes('jack','jack')
printNodes(head)
````

4-4. 노드 제거하는 함수 생성

````python
### 전역변수
memory = []
head, pre, current = None, None, None
dataArray = ['수영', '서현', '태연', '유리', '윤아'] # 실제 삽입될 데이터 목록 # 실무에서는 쿼리된 결과

### 함수
class Node():
    def __init__(self) :
        self.data = None
        self.link = None
        
# head를 집어넣으면 데이터를 연결시켜주는 함수       
def printNodes(start) :
    current = start  # 처음에는 head와 current가 같음
    if current == None: # current에 Node가 있으면 돌아감
        return 
    print(current.data, end = ' ') # head의 데이터 출력 
    while current.link != None : # current.link가 None을 잡지 않을 때까지 돌리기
        current = current.link # --> current = current.link = newNode 이니 결국 current가 다음 노드를 새로 갖게됐다고 생각하면 됨 
        print(current.data, end=' ') # --> 이 짓을 새로운 노드(current.link)가 None이 아닐때 까지 하면 됨
    print()
        

# 노드 삽입하는 함수
def insertNodes(findData, insertData) :
    global memory, head, pre, current
    # case 1 : head 앞에 노드 삽입하기 
    if (findData == head.data) :
        node = Node()
        node.data = insertData
        node.link = head # 새로운 노드의 link가 원래 head의 노드를 잡게하기
        head = node # 새로운 노드에 head 넘겨주기
        memory.append(node)
        return 
    # case 2 : 중간에 노드 삽입하기 
    current = head # 커렌트가 head에 있음을 확인 
    while (current.link != None) :
        pre = current # pre에 현재 노드 넘겨주기
        current = current.link # 현재 노드는 새로운 노드를 받기
        if current.data == findData :
            node = Node() # 삽입할 노드 생성
            node.data = insertData
            node.link = current
            pre.link = node
            memory.append(node) # 삽입된 노드 담기
            return
    # case 3 : 마지막에 노드 삽입하기 
    node = Node() # 마지막에 삽입할 노드 생성
    node.data = insertData
    current.link = node # 현재노드 링크가 새로운 노드 잡게하기
    return


# 노드 제거하는 함수 생성
def deleteNode(delData) :
    global memory, head, pre, current
    # case 1 : 삭제할 데이터가 head인 경우
    if (delData == head.data) :
        current = head # 현재 head를 없애기 위해 먼저 current에 복사해둠
        head = head.link # 그리고 나서 head는 다음 노드로 옮겨준다
        del(current) # 현재 head 제거 
        return
    
    # case 2 : 중간 데이터 삭제 
    current = head # current가 head에 있다고 확인시키기
    while (current.link != None) : # 삭제할 데이터를 찾을 때 까지 돌리기
        pre = current # pre를 먼저 현재 current로 옮기기
        current = current.link # current를 다음 노드로 옮기기
        if (current.data == delData) : # current의 data가 삭제할 데이터라면 (삭제할 데이터를 찾았다면)
            pre.link = current.link # 현재노드 앞의 link를 current가 잡고있는 다음 노드로 연결시켜주기
            del(current) # 찾은 데이터(=current) 삭제하기
            return
    
    return



### 메인 
# head를 만들기 위한 코드 
node = Node()
node.data = dataArray[0] # 첫 번째 노드
head = node # 첫 노드는 head 
memory.append(node) # 첫 노드를 memory에 넣음 

# 나머지는 몇 백개가 될지 모르니 for문으로 생성
for data in dataArray[1:] : # 두 번째 노드부터 끝까지 데이터를 삽입하기 위함 (0번째는 head가 있음)
    pre = node # pre : 앞 노드를 잡고있기 위한 변수
    node = Node() # 새로운 노드 생성
    node.data = data # 두 번째 노드에 데이터 삽입
    pre.link = node # 이전 노드의 링크가 새로운 노드를 잡게됨
    memory.append(node) # 새로 생긴 노드가 담기게됨

print('초기 상태')    
printNodes(head) # head를 집어넣으면 데이터를 연결시켜주는 함수 


# 삽입
# head에 새로운 node 삽입 
# print('\nhead에 삽입 :')
insertNodes('수영', 'melody')
# printNodes(head)

# 중간에 삽입하기 
# print('\n중간에 삽입 :')
insertNodes('태연', 'amy')
# printNodes(head)

# # 마지막에 삽입하기 
print('\n 데이터 삽입 :')
insertNodes('jack','jack')
printNodes(head)

# 삭제
print('\njack 삭제 :')
deleteNode('jack')
printNodes(head)

print('\namy 삭제 :')
deleteNode('amy')
printNodes(head)

print('\nmelody 삭제 :')
deleteNode('melody')
printNodes(head)
````
